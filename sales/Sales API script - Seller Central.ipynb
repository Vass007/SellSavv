```python
df_sales = pd.DataFrame()

for asin in asins:
    print(asin)
    response = sales.get_order_metrics(marketplaceIds=['A21TJRUUN4KGV'],interval=(start_time_str, end_time_str),asin=asin,granularity=Granularity.DAY)
    df_tmp = pd.DataFrame(response.payload)
    df_tmp['ASIN'] = asin
    df_sales = pd.concat([df_sales,df_tmp])
    time.sleep(2)
    
    
```
